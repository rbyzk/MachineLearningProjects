## 1. 🎯 BUSINESS CONTEXT & DECISION QUESTIONS

### 1.1 Why Telco Churn Matters
Customer churn is one of the most critical business risks for telecom operators because it directly
impacts recurring revenue, increases acquisition pressure, and weakens customer lifetime value.
Understanding churn is therefore essential to retain strategic customers and stabilize growth.

---

### 1.2 Analytical Objectives & Use Cases
This notebook aims to build a decision-ready churn intelligence framework that helps answer:
- Where does churn concentrate across lifecycle and contracts?
- Which customers combine risk + value exposure?
- What reasons and experience signals drive churn?
- Which regions demonstrate churn hot-spots?
- How can predictive models support smarter retention targeting?

Primary use cases:
- Retention strategy & prioritization
- Revenue-at-risk quantification
- Executive reporting & dashboards
- Marketing, CX and care operations enablement

---

### 1.3 Expected Business Impact
✔ Reduced churn through earlier detection  
✔ Better CLTV preservation  
✔ More targeted retention investments  
✔ Stronger alignment between analytics, business and operations

### 4.3 CLTV & Churn Score Exposure
We interpret churn and experience in the context of value exposure.

```python
def _range_sort_key(s):
    s = str(s)
    m = re.search(r"(\d+)", s)
    return int(m.group(1)) if m else 10**9

def _order_bins(series):
    vals = pd.Series(series.dropna().unique()).astype(str).tolist()
    return sorted(vals, key=_range_sort_key)

def _pretty_bin_label(s):
    s = str(s)
    nums = re.findall(r"-?\d+\.?\d*", s)
    if len(nums) >= 2:
        a, b = float(nums[0]), float(nums[1])
        if abs(a) >= 100 or abs(b) >= 100:
            return f"{a:.0f}–{b:.0f}"
        return f"{a:.1f}–{b:.1f}"
    s = s.replace("(", "").replace("]", "").replace("[", "").replace(")", "")
    s = re.sub(r"\s+", " ", s).strip()
    return s

def _fmt_money(x, _=None):
    x = float(x)
    if x >= 1_000_000:
        return f"${x/1_000_000:.1f}M"
    if x >= 1_000:
        return f"${x/1_000:.0f}K"
    return f"${x:.0f}"

need_cols = [c_churn_value, c_churn_score, c_cltv]
if any(c is None for c in need_cols):
    raise ValueError("Missing required columns for this chart: churn_value / churn_score / cltv")

plot_df = df[[c_churn_value, c_churn_score, c_cltv]].copy()
plot_df[c_churn_value] = plot_df[c_churn_value].astype(int)
plot_df[c_churn_score] = pd.to_numeric(plot_df[c_churn_score], errors="coerce")
plot_df[c_cltv] = pd.to_numeric(plot_df[c_cltv], errors="coerce")
plot_df = plot_df.dropna(subset=[c_churn_score, c_cltv]).copy()

risk_bin_col = "risk_bin"
value_bin_col = "value_bin"

if c_churn_score_cat is not None and c_churn_score_cat in df.columns:
    plot_df[risk_bin_col] = df.loc[plot_df.index, c_churn_score_cat].astype(str)
else:
    plot_df[risk_bin_col] = pd.qcut(plot_df[c_churn_score], 10, duplicates="drop").astype(str)

if c_cltv_cat is not None and c_cltv_cat in df.columns:
    plot_df[value_bin_col] = df.loc[plot_df.index, c_cltv_cat].astype(str)
else:
    plot_df[value_bin_col] = pd.qcut(plot_df[c_cltv], 10, duplicates="drop").astype(str)

risk_order = _order_bins(plot_df[risk_bin_col])
value_order = _order_bins(plot_df[value_bin_col])

cell = (
    plot_df
    .groupby([value_bin_col, risk_bin_col], observed=True)
    .agg(
        customers=(c_churn_value, "size"),
        churn_rate=(c_churn_value, "mean"),
        churned_cltv_exposure=(c_cltv, lambda s: float(s[plot_df.loc[s.index, c_churn_value] == 1].sum()))
    )
    .reset_index()
)

mat_exp = (
    cell.pivot(index=value_bin_col, columns=risk_bin_col, values="churned_cltv_exposure")
    .reindex(index=value_order, columns=risk_order)
    .fillna(0.0)
)

mat_rate = (
    cell.pivot(index=value_bin_col, columns=risk_bin_col, values="churn_rate")
    .reindex(index=value_order, columns=risk_order)
    .fillna(0.0)
)

mat_n = (
    cell.pivot(index=value_bin_col, columns=risk_bin_col, values="customers")
    .reindex(index=value_order, columns=risk_order)
    .fillna(0)
    .astype(int)
)

vmax = float(np.nanpercentile(mat_exp.values, 98)) if np.isfinite(mat_exp.values).any() else 1.0
vmax = max(vmax, 1.0)

n_rows, n_cols = mat_exp.shape
fig_w = max(11.6, 0.95 * n_cols + 6.2)
fig_h = max(6.1, 0.60 * n_rows + 3.0)

density = n_rows * n_cols
fs_main = float(np.clip(170 / max(n_rows, n_cols), 6.2, 9.4))
fs_sub = max(5.8, fs_main - 1.4)

show_detail = (density <= 72) and (min(n_rows, n_cols) <= 10)
show_thr = max(0.22 * vmax, float(np.nanpercentile(mat_exp.values, 70))) if np.isfinite(mat_exp.values).any() else 0.0

xlabels = [_pretty_bin_label(x) for x in mat_exp.columns.astype(str).tolist()]
ylabels = [_pretty_bin_label(y) for y in mat_exp.index.astype(str).tolist()]

fig, ax = plt.subplots(figsize=(fig_w, fig_h), dpi=160)

im = ax.imshow(mat_exp.values, cmap=telco_cmap_w, vmin=0.0, vmax=vmax, aspect="auto")

ax.set_xticks(range(n_cols))
ax.set_yticks(range(n_rows))
ax.set_xticklabels(xlabels, rotation=20, ha="right", fontsize=9, color=THEME["muted"])
ax.set_yticklabels(ylabels, fontsize=9, color=THEME["muted"])

ax.set_title("CLTV Exposure Heatmap — Risk (Churn Score) × Value (CLTV)", pad=14, fontweight="950")
ax.set_xlabel("Churn Score (Risk Buckets)")
ax.set_ylabel("CLTV (Value Buckets)")

ax.set_xticks(np.arange(-0.5, n_cols, 1), minor=True)
ax.set_yticks(np.arange(-0.5, n_rows, 1), minor=True)
ax.grid(which="minor", color=THEME["edge"], linewidth=1.0)
ax.tick_params(which="minor", bottom=False, left=False)

cbar = fig.colorbar(im, ax=ax, fraction=0.035, pad=0.03, format=FuncFormatter(_fmt_money))
cbar.outline.set_edgecolor(THEME["edge"])
cbar.outline.set_linewidth(1.0)
cbar.ax.set_ylabel("Churned CLTV Exposure (sum)", rotation=270, labelpad=18, color=THEME["muted"])
cbar.ax.tick_params(colors=THEME["muted"], length=0)

for i in range(n_rows):
    for j in range(n_cols):
        exp_v = float(mat_exp.values[i, j])
        rate_v = float(mat_rate.values[i, j]) * 100.0
        n_v = int(mat_n.values[i, j])

        if exp_v <= 0 and n_v <= 0:
            continue

        high = exp_v >= (0.55 * vmax)
        important = (exp_v >= show_thr)

        if show_detail and important:
            txt = f"{_fmt_money(exp_v)}\nCR {rate_v:.1f}% • n={n_v}"
            fsz = fs_main
        else:
            txt = f"{_fmt_money(exp_v)}"
            fsz = fs_sub

        ax.text(
            j, i, txt,
            ha="center", va="center",
            fontsize=fsz,
            fontweight="950" if (high or important) else "800",
            color="#FFFFFF" if high else THEME["slate"],
            linespacing=1.05
        )

for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
```

## 8. 🧭 DECISION SEGMENTATION — RISK × VALUE × SATISFACTION

### 8.1 Thresholding Strategy (Risk, Value, Satisfaction)
We build segmentation from:
- Risk = Churn Score  
- Value = CLTV  
- Experience = Satisfaction Score  

```python
seg = df.copy()

seg["risk_score"] = (
    seg[c_churn_score].astype(float) if c_churn_score is not None else np.nan
)
seg["value_score"] = (
    seg[c_cltv].astype(float) if c_cltv is not None else np.nan
)
seg["sat_score"] = (
    seg[c_satisfaction].astype(float) if c_satisfaction is not None else np.nan
)

risk_thr = (
    np.nanpercentile(seg["risk_score"], 70)
    if np.isfinite(seg["risk_score"]).any()
    else np.nan
)
value_thr = (
    np.nanpercentile(seg["value_score"], 70)
    if np.isfinite(seg["value_score"]).any()
    else np.nan
)

seg["risk_flag"] = np.where(
    np.isnan(seg["risk_score"]),
    0,
    (seg["risk_score"] >= risk_thr).astype(int),
)
seg["value_flag"] = np.where(
    np.isnan(seg["value_score"]),
    0,
    (seg["value_score"] >= value_thr).astype(int),
)
seg["low_sat_flag"] = np.where(
    np.isnan(seg["sat_score"]),
    0,
    (seg["sat_score"] <= 2).astype(int),
)

seg["retention_quadrant"] = np.select(
    [
        (seg["risk_flag"] == 1)
        & (seg["value_flag"] == 1)
        & (seg["low_sat_flag"] == 1),
        (seg["risk_flag"] == 1) & (seg["value_flag"] == 1),
        (seg["risk_flag"] == 1) & (seg["value_flag"] == 0),
        (seg["risk_flag"] == 0) & (seg["value_flag"] == 1),
    ],
    [
        "High Risk × High Value × Low Satisfaction",
        "High Risk × High Value",
        "High Risk × Lower Value",
        "Lower Risk × High Value",
    ],
    default="Lower Risk × Lower Value",
)

quad_df = (
    seg["retention_quadrant"]
      .value_counts()
      .rename_axis("retention_quadrant")
      .reset_index(name="customers")
)
quad_df["share"] = quad_df["customers"] / quad_df["customers"].sum()
quad_df["share_pct"] = (quad_df["share"] * 100).round(1)
quad_df = quad_df.sort_values("customers", ascending=False).reset_index(drop=True)
quad_df = quad_df[["retention_quadrant", "customers", "share_pct", "share"]]
```

### 8.2 Retention Quadrants & Shares

We define interpretable quadrants and compute their shares.

```python
def hex_to_rgba(h, a=0.10):
    h = str(h).lstrip("#")
    r = int(h[0:2], 16)
    g = int(h[2:4], 16)
    b = int(h[4:6], 16)
    return f"rgba({r},{g},{b},{a})"

row_palette = {
    "High Risk × High Value × Low Satisfaction": THEME["rose"],
    "High Risk × High Value": THEME["amber"],
    "High Risk × Lower Value": THEME["sky"],
    "Lower Risk × High Value": THEME["mint"],
    "Lower Risk × Lower Value": THEME["indigo"],
}

def row_style(row):
    c = row_palette.get(row["retention_quadrant"], THEME["gray"])
    bg = hex_to_rgba(c, 0.08)
    styles = [f"background-color: {bg};"] * len(row)
    styles[0] = f"background-color: {bg}; border-left: 6px solid {c};"
    return styles

sty = (
    quad_df.style
           .format({"customers": "{:,}",
                    "share_pct": "{:.1f}%",
                    "share": "{:.3f}"})
           .background_gradient(cmap=telco_cmap_w,
                                subset=["customers"],
                                axis=None)
           .bar(subset=["share_pct"],
                color=THEME["indigo"],
                vmin=0, vmax=100)
           .apply(row_style, axis=1)
)
sty = style_telco_table(sty)

html = f"""
<div style="width:100%; overflow-x:auto;">
  {sty.to_html()}
</div>
<style>
  table {{
    width: 100% !important;
    table-layout: auto !important;
  }}
</style>
"""
display(HTML(html))
```

## 9. 🤖 MODELLING — LEAKAGE-SAFE LEADERBOARD

### 9.1 Feature Set & Exclusions (no churn/CLTV indices)
We exclude leakage fields:
- Churn Score / Category
- CLTV / Category
- Churn Reason / Category
- Customer Status
- ZIP / Lat / Long
- Redundant churn labels



```python
from sklearn.model_selection import StratifiedKFold, cross_val_predict, RandomizedSearchCV
from sklearn.metrics import (
    roc_auc_score,
    average_precision_score,
    confusion_matrix,
    roc_curve,
    precision_recall_curve,
    auc,
)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    HistGradientBoostingClassifier,
)
from sklearn.svm import SVC
from scipy.stats import randint, uniform
from joblib import Memory

exclude = set([
    c_id,
    c_status,
    c_churn_label,
    c_churn_value,
    c_churn_score,
    c_churn_score_cat,
    c_cltv,
    c_cltv_cat,
    c_churn_cat,
    c_churn_reason,
    c_zip,
    c_lat,
    c_lon,
])
exclude = {x for x in exclude if x is not None}

X = df.drop(columns=[c for c in exclude if c in df.columns])
y = df[c_churn_value].astype(int)

cat_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()
num_cols = [c for c in X.columns if c not in cat_cols]

preprocess = ColumnTransformer(
    transformers=[
        ("num", Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler()),
        ]), num_cols),
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("ohe", OneHotEncoder(handle_unknown="ignore")),
        ]), cat_cols),
    ]
)

preprocess_dense = ColumnTransformer(
    transformers=[
        ("num", Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler()),
        ]), num_cols),
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("ohe", OneHotEncoder(handle_unknown="ignore")),
        ]), cat_cols),
    ],
    sparse_threshold=0.0,
)

def make_sample_weight(yv):
    yv = np.asarray(yv).astype(int)
    pos = (yv == 1).sum()
    neg = (yv == 0).sum()
    if pos == 0 or neg == 0:
        return np.ones_like(yv, dtype=float)
    w_pos = neg / pos
    return np.where(yv == 1, w_pos, 1.0).astype(float)
```

### 9.2 Model Zoo (LogReg, SVC, RF, ExtraTrees, HistGB)
We benchmark multiple models with cross-validated ROC-AUC.



```python
models = [
    (
        "LogReg",
        LogisticRegression(max_iter=5000, class_weight="balanced"),
        {"model__C": uniform(0.05, 10.0)},
    ),
    (
        "SVC-RBF",
        SVC(probability=True, class_weight="balanced", random_state=42),
        {
            "model__C": uniform(0.2, 20.0),
            "model__gamma": uniform(1e-4, 5e-2),
        },
    ),
    (
        "RandomForest",
        RandomForestClassifier(random_state=42, n_jobs=-1, class_weight="balanced"),
        {
            "model__n_estimators": randint(250, 800),
            "model__max_depth": randint(4, 20),
            "model__min_samples_split": randint(2, 14),
            "model__min_samples_leaf": randint(1, 8),
        },
    ),
    (
        "ExtraTrees",
        ExtraTreesClassifier(random_state=42, n_jobs=-1, class_weight="balanced"),
        {
            "model__n_estimators": randint(300, 900),
            "model__max_depth": randint(4, 22),
            "model__min_samples_split": randint(2, 14),
            "model__min_samples_leaf": randint(1, 8),
        },
    ),
    (
        "HistGB",
        HistGradientBoostingClassifier(random_state=42),
        {
            "model__learning_rate": uniform(0.02, 0.16),
            "model__max_depth": randint(2, 9),
            "model__max_leaf_nodes": randint(16, 72),
        },
    ),
]

memory = Memory(location="/kaggle/working/sk_cache", verbose=0)
cv_fast = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
FAST_N_ITER = 12

rows = []
best_name, best_pipe, best_auc = None, None, -1.0

for name, estimator, space in models:
    pp = preprocess_dense if name == "HistGB" else preprocess
    pipe = Pipeline([("preprocess", pp), ("model", estimator)], memory=memory)

    search = RandomizedSearchCV(
        pipe,
        param_distributions=space,
        n_iter=FAST_N_ITER,
        scoring="roc_auc",
        cv=cv_fast,
        n_jobs=-1,
        random_state=42,
        verbose=0,
    )

    if name == "HistGB":
        sw = make_sample_weight(y)
        search.fit(X, y, model__sample_weight=sw)
    else:
        search.fit(X, y)

    auc_cv = float(search.best_score_)
    rows.append({"Model": name, "ROC_AUC_CV": auc_cv})

    if auc_cv > best_auc:
        best_auc = auc_cv
        best_name = name
        best_pipe = search.best_estimator_

scores_df = (
    pd.DataFrame(rows)
      .sort_values("ROC_AUC_CV", ascending=False)
      .reset_index(drop=True)
)

scores_df, best_name, best_auc
```

### 9.3 ROC-AUC Leaderboard (FAST CV)

We use a fast randomized search to derive a model leaderboard.

```python
plot_df = scores_df.sort_values("ROC_AUC_CV", ascending=True).reset_index(drop=True)

fig, ax = plt.subplots(figsize=(9.4, 4.8), dpi=140)
ax.barh(
    plot_df["Model"].astype(str),
    plot_df["ROC_AUC_CV"].astype(float),
    height=0.62,
    color=THEME["indigo"],
    alpha=0.90,
    edgecolor="none",
)

for i, v in enumerate(plot_df["ROC_AUC_CV"].astype(float).values):
    ax.text(
        v + 0.004, i,
        f"{v:.3f}",
        va="center",
        fontsize=10,
        fontweight=950,
        color=THEME["ink"],
    )

ax.set_xlim(
    max(0.65, float(plot_df["ROC_AUC_CV"].min()) - 0.03),
    min(1.0, float(plot_df["ROC_AUC_CV"].max()) + 0.06),
)
ax.set_xlabel("ROC-AUC (Cross-Validated)")
ax.set_title("Model Leaderboard — Leakage-Safe CV (FAST)",
             pad=12, fontweight="950")
ax_minimal(ax, grid_axis="x")
fig.subplots_adjust(left=0.28, right=0.98, top=0.88, bottom=0.16)
plt.show()

best_name, best_auc
```

## 12. 🚀 CONCLUSION & NEXT STEPS

### 12.1 What We Learned
- Churn concentrates in specific lifecycle bands and payment behaviors.  
- Experience deficits (low satisfaction) align strongly with churn.  
- Risk is meaningful only when combined with value (CLTV).  
- Geo and churn category views provide operational prioritization.

---

### 12.2 Retention Playbook Directions
- **High Risk × High Value × Low Satisfaction** → immediate outreach, service recovery, premium support.  
- **High Risk × High Value** → contract migration, loyalty offers, proactive care.  
- **High Risk × Lower Value** → low-cost digital nudges, friction reduction (billing/payment).  
- **Lower Risk × High Value** → white-glove retention, satisfaction monitoring, cross-sell.  
- **Lower Risk × Lower Value** → efficient lifecycle communications.

---

### 12.3 Future Enhancements & Experiments
- Run controlled retention experiments (A/B) by quadrant.  
- Track churn, satisfaction and revenue uplift over time.
